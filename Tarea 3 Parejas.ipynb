**2. Haga una funcion en Python que dado un numero entre 1 y 20. Y un caracter cualquiera despliegue cuatro patrones triangulares.**

```python
def triangulo(a,b):
    if(a>0 and a<21):
      for i in range(1,a + 1, 1):
        resultado = b * i
        print("\n " + (resultado))
      print("-----------------------------------------------------")
      for i in range(a,-1,-1):
        resultado = b * i
        print("\n " + resultado)
      print("-----------------------------------------------------")
      for i in range(a,-1,-1):
        resultado = (" " * (a - i - 1) + b * (2 * i + 1))
        print(resultado)
      print("-----------------------------------------------------")
      for i in range(1,a+1,1):
        resultado = (" " * (a - i - 1) + b * (2 * i + 1))
        print(resultado)
      print("-----------------------------------------------------")
    else:
      print("Ingrese datos correctos, por favor.")
    
def main():
    try:
      entradaNum = int(input("Ingrese un numero de 1 al 20: "))
      entradaSim = input("Ingrese un simbolo: ")
      
      triangulo(entradaNum,entradaSim)
      
    except(ValueError):
      print("Ingrese datos correctos, por favor.")

main()
```

**1. Escriba una funcion en Python que dada una tupla de elementos (numeros o letras) esta se convierta a un string. Tomar nota del siguiente ejemplo:**
* Si la entrada fuera: (10, 20, 40, 5, 70)
* La salida deberıa ser: 102040570

```python
def ConvertirString(tup): 
  str =  ''.join(tup)
  return str
  

def main():
    try: 
      d1 = input("Ingrese primer elemento: ") 
      d2 = input("Ingrese segundo elemento: ") 
      d3 = input("Ingrese tercer elemento: ") 
      d4 = input("Ingrese cuarto elemento: ") 
      d5 = input("Ingrese quinto elemento: ")  

      tuple = (d1, d2,d3,d4,d5)
      str = ConvertirString(tuple)
      print("El resultado en string de la tupla es: ", str) 
    except(ValueError):
      print("Ingrese datos correctos, por favor.")

main()
```

**2. Cree un programa en Python que dada una lista de tuplas, elimine las tuplas que estan vacıas.
Tomar nota del siguiente ejemplo:**
* Si la lista fuera: [(), (), (’X’,), (’a’, ’b’), (’a’, ’b’, ’c’), (’d’)]
* El resultado deber´ıa ser: [(’X’), (’a’, ’b’), (’a’, ’b’, ’c’), ’d’]


```python
def elimina_vacio(lista):
    nueva = []
    for dato in lista:
        if dato != '()':
            nueva.append(dato)
    return nueva

def main():
    try: 
      d1 = input("Ingrese primer elemento: ") 
      d2 = input("Ingrese segundo elemento: ") 
      d3 = input("Ingrese tercer elemento: ") 
      d4 = input("Ingrese cuarto elemento: ") 
      d5 = input("Ingrese quinto elemento: ")  

      lista = (d1, d2,d3,d4,d5)
      lista_nueva = elimina_vacio(lista)
      print("El resultado en la lista sin elementos vacio es: ", lista_nueva)
 
    except(ValueError):
      print("Ingrese datos correctos, por favor.")

main()
```
